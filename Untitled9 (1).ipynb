```python
# For data handling
import pandas as pd
import re
import numpy as np

# For machine learning
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Replace the path with your CSV file path
df = pd.read_csv("deeplearning.csv")

# See first 5 rows to understand data
print(df.head())
```

```python
# Usually text is in a column with 'text' or 'tweet'
text_col = [c for c in df.columns if 'text' in c.lower() or 'tweet' in c.lower()][0]

# Usually label is in a column called 'label', 'category' or 'sentiment'
label_col = [c for c in df.columns if c.lower() in ('label', 'category', 'sentiment')][0]

print("Text column:", text_col)
print("Label column:", label_col)
```

```python
def map_label(v):
    # Convert 0/1/-1 to Neutral/Positive/Negative
    return {0:"Neutral", 1:"Positive", -1:"Negative"}.get(v, str(v))

df['label_str'] = df[label_col].apply(map_label)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download("wordnet")
nltk.download("omw-1.4")
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

lemmatizer = WordNetLemmatizer()
stopwords = set(ENGLISH_STOP_WORDS)



def clean_text(s):
    if pd.isna(s):   # handle NaN directly
        return ""
    s = str(s)
    s = re.sub(r"http\S+", " ", s)   # remove URLs
    s = re.sub(r"@\w+", " ", s)      # remove mentions
    s = re.sub(r"#\w+", " ", s)      # remove hashtags
    s = re.sub(r'[^A-Za-z0-9\s]', ' ', s)  # remove special characters
    s = s.lower()
    s = " ".join([lemmatizer.lemmatize(w) for w in s.split() if w not in stopwords])  # remove stopwords + lemmatize
    return s

# Apply cleaning
df['clean_text'] = df[text_col].apply(clean_text)
```

```python
X = df['clean_text']      # features
y = df['label_str']       # target labels

# Encode labels to numbers
le = LabelEncoder()
y_num = le.fit_transform(y)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# 🔹 Class distribution plot
plt.figure(figsize=(6,4))
sns.countplot(x='label_str', data=df, palette="Set2")
plt.title("Sentiment Class Distribution")
plt.show()

# 🔹 WordCloud for each sentiment
for sentiment in df['label_str'].unique():
    texts = df.loc[df['label_str'] == sentiment, 'clean_text'].dropna().astype(str)
    text = " ".join(texts)

    wc = WordCloud(width=800, height=400, background_color="white").generate(text)
    plt.figure(figsize=(10,5))
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(f"WordCloud - {sentiment}")
    plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_num, test_size=0.2, random_state=42, stratify=y_num
)
```

```python
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
clf = MultinomialNB()
clf.fit(X_train_tfidf, y_train)

# Make predictions
preds = clf.predict(X_test_tfidf)

# Convert numbers back to original labels
pred_labels = le.inverse_transform(preds)
true_labels = le.inverse_transform(y_test)
```

```python
print(classification_report(true_labels, pred_labels))
print("Accuracy:", accuracy_score(true_labels, pred_labels))
```

Overall Performance

Accuracy: ~71%
→ The model correctly predicted about 71 out of 100 tweets.

Weighted Average F1: 0.71
→ Since the dataset is imbalanced (more positives & neutrals than negatives), the weighted average is a better overall indicator than the macro average. A score of 0.71 is decent for sentiment analysis.

 Per-Class Performance

Negative (7102 samples)

Precision: 0.82 → When the model predicts Negative, it’s right 82% of the time (very good).

Recall: 0.47 → Out of all actual Negative tweets, it catches only 47%.

F1: 0.60 → Balanced score is moderate.
 Interpretation: Model is cautious about labeling tweets Negative (high precision, but misses many Negatives → low recall).

Neutral (11043 samples)

Precision: 0.76 → Fairly accurate when it predicts Neutral.

Recall: 0.67 → Captures 67% of actual Neutral tweets.

F1: 0.72 → Solid balance.
 Interpretation: Neutral class is handled well and balanced.

Positive (14450 samples)

Precision: 0.67 → When model predicts Positive, only 67% are truly Positive (so, some False Positives).

Recall: 0.87 → It successfully finds 87% of all actual Positives.

F1: 0.75 → Strong performance overall.
 Interpretation: Model is very good at catching Positive tweets (high recall), but sometimes over-predicts them.



```python
try:
    import tensorflow as tf
    from tensorflow.keras.preprocessing.text import Tokenizer
    from tensorflow.keras.preprocessing.sequence import pad_sequences
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, Bidirectional
    from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
    from sklearn.metrics import classification_report, accuracy_score

    # Prepare tokenizer
    max_words = 10000
    max_len = 50
    embedding_dim = 100

    tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>")
    tokenizer.fit_on_texts(X_train)
    X_train_seq = pad_sequences(tokenizer.texts_to_sequences(X_train), maxlen=max_len)
    X_test_seq = pad_sequences(tokenizer.texts_to_sequences(X_test), maxlen=max_len)

    # Build LSTM model
    model = Sequential([
        Embedding(max_words, embedding_dim, input_length=max_len),
        Bidirectional(LSTM(64)),
        Dropout(0.5),
        Dense(32, activation='relu'),
        Dense(len(le.classes_), activation='softmax')
    ])

    model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

    # Callbacks
    callbacks = [
        EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True),
        ModelCheckpoint("best_model.h5", save_best_only=True)
    ]

    # Train
    history = model.fit(
        X_train_seq, y_train,
        epochs=10, batch_size=32,
        validation_split=0.1,
        callbacks=callbacks
    )

    # Predict
    preds_idx = model.predict(X_test_seq).argmax(axis=1)
    print(classification_report(le.inverse_transform(y_test), le.inverse_transform(preds_idx)))
    print("LSTM Accuracy:", accuracy_score(le.inverse_transform(y_test), le.inverse_transform(preds_idx)))

except Exception as e:
    print("LSTM skipped:", e)
```

The BiLSTM model achieved an overall accuracy of 86.7%, a significant improvement over classical approaches (71%). It performs best on Neutral and Positive tweets, with F1-scores of 0.89 and 0.87 respectively, while Negative tweets achieved an F1-score of 0.82. Validation accuracy peaked at ~88% by the 2nd–3rd epoch before showing mild overfitting. This demonstrates that deep learning with LSTMs is highly effective for sentiment classification tasks.

```python
# Training curves
plt.figure(figsize=(8,4))
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title("Model Accuracy")
plt.legend()
plt.show()

plt.figure(figsize=(8,4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("Model Loss")
plt.legend()
plt.show()

# Confusion Matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(le.inverse_transform(y_test), le.inverse_transform(preds_idx), labels=le.classes_)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```
